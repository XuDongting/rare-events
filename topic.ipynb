```python
# Do hypothesis analysis on the data. The t-test is about mean and f-test is about variance.
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ttest_ind   
pvals= dict([])
for item in sensortype:
    item_data = data.loc[:,item]
    print(data.loc[:,item].head(3))
    item_1 = data.loc[data.y==1,item]
    item_0 = data.loc[data.y==0,item]     
    t_result = ttest_ind(item_0, item_1)        
    print('t test of', item, t_result)
    pvals[item]= t_result[1]
sort_pvals = sorted(pvals.items(), key = lambda kv: kv[1])
print(sort_pvals)   

```
